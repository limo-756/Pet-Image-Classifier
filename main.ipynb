```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("bhavikjikadara/dog-and-cat-classification-dataset")

print("Path to dataset files:", path)
```

```python
from pathlib import Path
from random import shuffle
import pandas as pd
import os

classes = ['cat', 'dog']

train_df = pd.DataFrame(columns=['image', 'label', 'encoded_label'])
val_df = pd.DataFrame(columns=['image', 'label', 'encoded_label'])
test_df = pd.DataFrame(columns=['image', 'label', 'encoded_label'])
base_dir = "./dataset"

for idx, class_name in enumerate(classes):
    path = os.path.join(base_dir, class_name)
    images = os.listdir(path)
    print(f"Found {len(images)} images for class: {class_name}")
    shuffle(images)
    train_ending = int(0.8 * len(images))
    val_ending = train_ending + int(0.1 * len(images))

    rows = []
    for image in images[:train_ending]:
        rows.append({'image': os.path.join(path, image), 'label': class_name, 'encoded_label': idx})
    df = pd.DataFrame(columns=['image', 'label', 'encoded_label'], data=rows)
    train_df = pd.concat([df, train_df])

    rows = []
    for image in images[train_ending: val_ending]:
        rows.append({'image': os.path.join(path, image), 'label': class_name, 'encoded_label': idx})
    df = pd.DataFrame(columns=['image', 'label', 'encoded_label'], data=rows)
    val_df = pd.concat([df, val_df])

    rows = []
    for image in images[val_ending:]:
        rows.append({'image': os.path.join(path, image), 'label': class_name, 'encoded_label': idx})
    df = pd.DataFrame(columns=['image', 'label', 'encoded_label'], data=rows)
    test_df = pd.concat([df, test_df])

train_df.reset_index(drop=True, inplace=True)
val_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
train_df.index.name = 'index'
val_df.index.name = 'index'
test_df.index.name = 'index'
```
